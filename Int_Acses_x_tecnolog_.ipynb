```python
df = pd.read_csv("data_original\Internet_Accesos-por-tecnologia.csv")
df #vusualizacion de dataframe
```

```python
df.info()#visualizar cantidad de columnas, filas tipo de datos
```

 cambios de tipo de tipo de datos de columna

```python
df['ADSL'] = df['ADSL'].astype(str).str.replace('.', '').astype(float)
df['Cablemodem'] = df['Cablemodem'].astype(str).str.replace('.', '').astype(float)
df['Fibra óptica'] = df['Fibra óptica'].astype(str).str.replace('.', '').astype(float)
df['Total'] = df['Total'].astype(str).str.replace('.', '').astype(float)
df['Trimestre'] = df['Trimestre'].astype('category')
df['Año'] = df['Año'].astype('str')
```

```python
df.info()#corroborar cambios
```

```python
df.describe()#resumen estadistico
```
