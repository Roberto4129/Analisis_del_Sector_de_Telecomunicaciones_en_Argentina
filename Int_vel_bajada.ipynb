```python
from ydata_profiling import ProfileReport
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Int_vel_bajada

Velocidad Media de bajada de Internet fijo
Serie histórica de la velocidad media de descarga de Internet nacional (trimestral)

Visualizar dataframe

```python
df = pd.read_csv("data_original\historico_velocidad_internet.csv")
df
```

```python
df.info()#visualizar cantidad de columnas, filas tipo de datos
```

cambio de tipo de dato en las columnas

```python
# Remover los puntos de la columna 
df['Mbps (Media de bajada)'] = df['Mbps (Media de bajada)'].str.replace(',', '.')
```

```python
# cambios de tipo de datos de columna
df['Mbps (Media de bajada)'] = df['Mbps (Media de bajada)'].astype(float)
df['Trimestre'] = df['Trimestre'].astype('category')
df['Año'] = df['Año'].astype('str')
```

```python
df.describe()#resumen estadistico
```

```python
sns.distplot(df['Mbps (Media de bajada)'], hist=True, kde=True, rug=True)
```

_La velocidad media de baja rondan ente 5 y 60 Mbps aprox.

_Alta densidad entre 5 y 20 Mbps

```python
fig=plt.figure(figsize=(20,7))

ax1=fig.add_subplot(131)
#ax2=fig.add_subplot(132)

sns.boxplot(data=df,y='Mbps (Media de bajada)',ax=ax1);
#sns.boxplot(data=df,y='Dial up',ax=ax2);
```

_No se observan outliers

_El 75%  de bajada ronda en 40Mbps

_La mediana esta alrededor de 15 Mbps

```python
ax = sns.scatterplot(x="Mbps (Media de bajada)", y="Año", data=df)
```

Desde el año 2021 hasta el utlmos registro 2022, se ven registro desde 45 Mbps hasta llegar su maxiom de 62.46 Mbps

Entre el 2019 y el 2020 se ve un salto de promedio de bajada de 30 a 40 Mbps
