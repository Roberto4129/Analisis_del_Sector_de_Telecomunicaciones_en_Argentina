```python
# Contar la frecuencia de cada valor en la columna 
frecuencia = df['SATELITAL'].value_counts()

# Calcular el porcentaje de cada valor
porcentaje = frecuencia / len(df) * 100

# Crear el gráfico de barras
frecuencia.plot(kind='bar', color='skyblue')

# Agregar etiquetas y título
plt.xlabel('Valor de SATELITALS')
plt.ylabel('Frecuencia')
plt.title('Frecuencia y Porcentaje de valores en la columna SATELITAL')

# Mostrar la frecuencia y el porcentaje en cada barra
for i, v in enumerate(frecuencia):
    plt.text(i, v + 0.5, f'{v} ({porcentaje[i]:.1f}%)', ha='center')

# Mostrar el gráfico de barras
plt.show()
```

Presencia del 25% de servicio SATELITAL
